```python
    from selenium import webdriver
    from selenium.webdriver.common.by import By
    from selenium.webdriver.support.ui import WebDriverWait
    from selenium.webdriver.support import expected_conditions as EC
    from selenium.webdriver.common.action_chains import ActionChains
    from selenium.webdriver.common.keys import Keys
    
    # specify the url of the business page on Google
    url = 'https://www.google.com/maps/place/Tata+Institute+of+Social+Sciences/@19.0446964,72.912484,15z/data=!4m8!3m7!1s0x3be7c608f591996b:0x69dbb7e2f00ee03c!8m2!3d19.0446964!4d72.912484!9m1!1b1!16s%2Fm%2F027kbf7?entry=ttu'

    # Create an instance of the Chrome WebDriver
    driver = webdriver.Chrome()

    # navigate to the specified url
    driver.get(url)

    totalRev = "div div.fontBodySmall"
    username = ".d4r55"
    reviews = "wiI7pd"

    wait = WebDriverWait(driver, 40)

    totalRevCount = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, totalRev))).get_attribute("textContent").split(' ')[0].replace(',','').replace('.','')
    print("totalRevCount - ", totalRevCount)
    
    for i in range(8):
            ActionChains(driver).key_down(Keys.ARROW_DOWN).perform()
    #wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, totalRev))).click()

    mydict = {}
    found = 0

    while found < int(totalRevCount):

        review_elements = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, reviews)))
        #review_elements = "Good"
        reviewer_names = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, username)))
        #print("completed",reviewer_names)

        found = len(mydict)
        for rev, name in zip(review_elements, reviewer_names):
            mydict[name.text] = rev.text
            if len(rev.text) == 0:
                found = int(totalRevCount) + 1
                break

        for i in range(8):
            ActionChains(driver).key_down(Keys.ARROW_DOWN).perform()

        print("found - ", found)

        print(mydict)

       # time.sleep(2)
```

```python
# Your existing code to populate the 'mydict' dictionary

# Specify the file name to which you want to export the data
file_name = "reviews.txt"

# Open the file in write mode with UTF-8 encoding
with open(file_name, "w", encoding="utf-8") as file:
    for name, review in mydict.items():
        # Write each reviewer name and review to the file
        file.write(f"Reviewer: {name}\n")
        file.write(f"Review: {review}\n\n")

print(f"Data exported to {file_name}")
```

```python
import csv

# Your existing code to populate the 'mydict' dictionary

# Specify the file name for the CSV output
csv_file_name = "reviews.csv"

# Create a list to hold the data
data = []
for name, review in mydict.items():
    data.append([name, review])

# Open the CSV file in write mode
with open(csv_file_name, "w", newline="", encoding="utf-8") as csv_file:
    csv_writer = csv.writer(csv_file)
    
    # Write the header row (if needed)
    csv_writer.writerow(["Reviewer", "Review"])
    
    # Write the data to the CSV file
    csv_writer.writerows(data)

print(f"Data exported to {csv_file_name}")
```
